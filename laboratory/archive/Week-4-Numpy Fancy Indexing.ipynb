```python
# Write Program using Numpy fancy Indexing example: Selecting random points.
import numpy as np
import matplotlib.pyplot as plt
rand = np.random.RandomState(42)

x = rand.randint(100, size=10)
x
```

```python
# individual retrieval of elements
print(x[3], x[7], x[-2], x[-8], x[1:6:2], x[-2:-6:-2])
```

```python
print(x)
ind=[2,4,6]
print(x[ind])
print("===========OR=============")
print(x[[2,4,6]])
```

```python
# Fancy indexing - 2D
print(x)
ind = np.array([[3, 7],
                [4, 5]])
print(x[ind])
print("===========OR=============")
print(x[ [[3, 7],[4, 5]] ])
```

```python
# Fancy Indexing - 2D
print(x)
ind = np.array([[3, 7],
                [4, 5],
                [6, 8]])
x[ind]
```

```python
# Generating a array of shape (100,2) i.e., 100 rows and 2 columns. each column represents one feature/variable.
# 1st element of each row represents variable1 values.
# 2nd element of each row represents variable2 values.
import numpy as np
# Create a random number generator with a fixed seed for reproducible results
rand = np.random.RandomState(42)
# Display the RandomState object information
print(rand)
# Define the mean vector for a 2-dimensional normal distribution
mean = [0, 0]
# Define the covariance matrix showing variance and correlation between variables 1 and 2
cov = [[1, 2],
 [2, 5]]
# Generate 100 samples from a 2D multivariate normal (Gaussian) distribution
# The values generated must satisfy above mean and covariance matrix given above
X = rand.multivariate_normal(mean, cov, 100)
# Print the generated random samples
print(X)
# Print the shape of the generated array (rows = samples, columns = features)
print(X.shape)
```

```python
# Randomly select 20 unique row indices from the total number of rows in X
indices = np.random.choice(X.shape[0], 20, replace=False)
# Print the randomly selected indices
print(indices)
# Select rows from X using the chosen indices (fancy indexing)
selection = X[indices] # fancy indexing here
# Print the shape of the selected subset (20 rows, same number of columns)
print(selection.shape)
```

```python
# Plot all data points using the first column as X-axis and
# second column as Y-axis with transparency
plt.scatter(X[:, 0], X[:, 1], alpha=0.5)
# Plot the selected subset of points on top with black color and larger marker size
plt.scatter(selection[:, 0], selection[:, 1],
 facecolor='black', s=90)
```
